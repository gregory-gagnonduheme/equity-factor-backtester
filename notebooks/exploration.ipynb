```python
# Hard-coded ticket list (30x) to get started; top-30 S&P500 components by weight (as of June 9th 2026)
TICKERS = ['NVDA', 'AAPL', 'MSFT', 'AMZN', 'GOOGL',
           'GOOG', 'AVGO', 'TSLA', 'META', 'BRK-B',
           'MU', 'LLY', 'WMT', 'JPM', 'AMD',
           'XOM', 'V', 'ORCL', 'JNJ', 'INTC', 
           'CSCO', 'MA', 'COST', 'CAT', 'LRCX',
           'AMAT', 'ABBV', 'BAC', 'UNH', 'CVX']
```

```python
tickers = yf.Tickers(TICKERS)
```

```python
data = yf.download(TICKERS, period='10y', interval='1d', auto_adjust=False)
```
